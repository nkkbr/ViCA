```python
import json
```

```python
merged_json_list = []
for json_file in json_files:
    with open(json_file,'r') as f:
        data_list_curr = json.load(f)
    merged_json_list += data_list_curr
```

```python
arkitscenes_video_set = set()
for data in merged_json_list:
    if data['video'].startswith("arkitscenes/"):
        arkitscenes_video_set.add(data['video'].split('arkitscenes/')[1].split('.mp4')[0])

scannet_video_set = set()
for data in merged_json_list:
    if data['video'].startswith("scannet/"):
        scannet_video_set.add(data['video'].split('scannet/')[1].split('.mp4')[0])

scannetpp_video_set = set()
for data in merged_json_list:
    if data['video'].startswith("scannetpp/"):
        scannetpp_video_set.add(data['video'].split('scannetpp/')[1].split('.mp4')[0])
```
